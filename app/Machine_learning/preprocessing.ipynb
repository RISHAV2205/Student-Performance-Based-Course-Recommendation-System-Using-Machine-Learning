```python
import pandas as pd
data=pd.read_csv(r"D:\academics\project\Performance_Based_Recommendation\app\Data\student_data.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.head()
```

```python
data["performance_score"] = (data["G1"] + data["G2"] + data["G3"]) / 3
```

creating performance category

```python
def performance_label(score):
    if score < 10:
        return "Weak"
    elif score < 15:
        return "Average"
    else:
        return "Strong"
data["performance"] = data["performance_score"].apply(performance_label)
```

feature selection:

```python
X = data.drop(["G1","G2","G3","performance","performance_score"], axis=1)
y = data["performance"]
```

Feature scaling:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Splitting Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test)
y_pred_dt = dt_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_lr = accuracy_score(y_test, y_pred_lr)
acc_dt = accuracy_score(y_test, y_pred_dt)
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Logistic Regression Accuracy:", acc_lr)
print("Decision Tree Accuracy:", acc_dt)
print("Random Forest Accuracy:", acc_rf)
```

Saving the trained model

```python
import joblib
joblib.dump(lr_model, "student_model.pkl")
```

Recommendation Engine

```python
def recommend_courses(student):
    recommendations = []
    if student["studytime"] <= 2:
        recommendations.append("Study Planning and Learning Strategies")
    if student["failures"] >= 1:
        recommendations.append("Remedial Academic Support Program")
    if student["absences"] > 10:
        recommendations.append("Attendance and Time Management Training")
    if student["goout"] >= 4:
        recommendations.append("Focus and Productivity Development")
    if student["G3"] >= 15:
        recommendations.append("Advanced Learning Program")
    elif student["G3"] >= 10:
        recommendations.append("Intermediate Skill Development Courses")
    else:
        recommendations.append("Foundation Level Academic Support")
    return list(set(recommendations))
```

```python
student = data.iloc[0]

courses = recommend_courses(student)
print("Recommended Courses:")
for c in courses:
    print("-", c)
```
