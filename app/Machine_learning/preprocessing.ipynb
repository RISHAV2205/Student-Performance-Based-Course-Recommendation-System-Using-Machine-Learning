```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.head()
```

```python
data["performance_score"] = (data["G1"] + data["G2"] + data["G3"]) / 3
```

creating performance category

```python
def performance_label(score):
    if score < 10:
        return "Weak"
    elif score < 15:
        return "Average"
    else:
        return "Strong"
data["performance"] = data["performance_score"].apply(performance_label)
```

feature selection:

```python
X = data.drop(["G1","G2","G3","performance","performance_score"], axis=1)
y = data["performance"]
```

Feature scaling:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Splitting Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```
